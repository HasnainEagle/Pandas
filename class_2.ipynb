```python
# find all the final winners 

# Answer

# first we find all final matches 
mask = ipl["MatchNumber"] == "Final"
final_matches = ipl[mask]

# select winning teams 
Final_Winners = final_matches.loc[0:,["Season","WinningTeam"]]
# Final_Winners 
```

```python
# how many super over finishes have  occured

# Answer 

mask = ipl["SuperOver"].values == "Y"
super_over = ipl[mask]["SuperOver"].count()
# super_over
```

```python
# how many matches has csk won in kolkata 
matches_in_kolkata = ipl[ipl["City"] == "Kolkata"]
bool_csk_matches = matches_in_kolkata["WinningTeam"].values == "Chennai Super Kings"
csk_win_at_kolkata = bool_csk_matches[bool_csk_matches].sum()
# csk_win_at_kolkata

# second logic 
csk_win_at_kolkata = ipl[(ipl["City"] == "Kolkata") &  (ipl["WinningTeam"] == "Chennai Super Kings")].shape[0]
# csk_win_at_kolkata
```

```python
# toss winner is match winner in percentage
toss_winner_is_match_winner = ((ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0])/ipl.shape[0])*100
toss_winner_is_match_winner
```
