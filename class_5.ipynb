```python
# find ranking of each movie in the group according to IMDB score.
```

```python
def rank_each_movie(group):
    group["Ranking"] = group["IMDB_Rating"].rank(ascending=False)
    return group
```

```python
genres.apply(rank_each_movie)
```

```python
# find normalized IMDB rating group wise
```

```python
def normalized_rating(group):
    group["Noramlized_Rating"] = group["IMDB_Rating"]-group["IMDB_Rating"].min()/group["IMDB_Rating"].max()-group["IMDB_Rating"].min()
    return group
```

```python
genres.apply(normalized_rating)
```

```python
# IPL Dataset
ipl = pd.read_csv("deliveries.csv")
ipl.head()
```

```python
ipl.columns
```

```python
ipl.shape
```

```python
# find the top 10 batsman in terms of runs
batsman = ipl.groupby("batsman")
# here we use reset_index() to convert the (series into DataFrame)
batsman["batsman_runs"].sum().sort_values(ascending=False).reset_index().head(10)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl["batsman_runs"] == 6]
batsman_hit_six = six.groupby("batsman")
# batsman_hit_six.count().sort_values("batsman_runs",ascending=False)["batsman_runs"].head(1).index[0]  #not prefer
# Another way
batsman_hit_six["batsman"].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
over_above_15_df = ipl[ipl["over"] > 15]
batsman_with_6_or_4 = over_above_15_df[(over_above_15_df["batsman_runs"] == 4) | (over_above_15_df["batsman_runs"] == 6)]
batsman_with_6_or_4.groupby("batsman")["batsman"].count().sort_values(ascending=False).head(1)
```

```python
# find V Kohli's record against all teams
def kohli_record(group):
    spec_col = group.agg(
        {
            "batsman_runs":"sum",
            "legbye_runs":"sum"
        }
    )
    return spec_col
```

```python
batsman_Kohli = ipl[ipl["batsman"] == "V Kohli"]
bowling_team = batsman_Kohli.groupby("bowling_team")
batsman_runs_and_legbye_runs_df = bowling_team.apply(kohli_record)
batsman_runs_and_legbye_runs_df
```

```python
batsman_runs_and_legbye_runs_df.plot(kind="bar",color=["green","red"])
```

```python
# Add some more records
```

```python
kohli_4 = ipl[(ipl["batsman_runs"] == 4) & (ipl["batsman"] == "V Kohli")]
kohli_6 = ipl[(ipl["batsman_runs"] == 6) & (ipl["batsman"] == "V Kohli")]
fours_by_kohli = kohli_4.groupby("bowling_team")
sixes_by_kohli = kohli_6.groupby("bowling_team")
```

```python
four_df = fours_by_kohli["batsman_runs"].count().reset_index()
sixes_df = sixes_by_kohli["batsman_runs"].count().reset_index()
```

```python
# Renaming the columns
four_df.rename(columns={"batsman_runs":"No_of_Fours"},inplace=True)
sixes_df.rename(columns={"batsman_runs":"No_of_Sixes"},inplace=True)
```

```python
# concate two dataframes
four_and_sixes = pd.concat([four_df,sixes_df],axis=1)
# removing duplicate columns
four_and_sixes_df = four_and_sixes.loc[:,~four_and_sixes.columns.duplicated()]
four_and_sixes_df.set_index("bowling_team",inplace=True)
```

```python
kohli_records_against_all_teams_df = pd.concat([four_and_sixes_df,batsman_runs_and_legbye_runs_df],axis=1)
kohli_records_against_all_teams_df
```

```python
# plotting
ax = kohli_records_against_all_teams_df.plot(kind='bar', figsize=(10, 6))
ax.set_xlabel("Teams")
ax.set_ylabel("Counts")
ax.set_title("Virat Kohli Record Against All Ipl Teams")
```

```python
# Create a function that can return the total runs of any batsman
```

```python
def batsman_total_runs(batsman_name):
    highest_runs = ipl.groupby("batsman").get_group(batsman_name)["batsman_runs"].sum()
    print(f"{batsman_name} total runs are {highest_runs}")
```

```python
batsman_total_runs("V Kohli")
```

```python
# Create a function that can return the highest score of any batsman.
```

```python
def batsman_highest_runs(batsman_name):
    temp = ipl[ipl["batsman"] == batsman_name]
    res = temp.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).head(1).values[0]
    print(f"highest runs of {batsman_name} is {res}")
```

```python
batsman_highest_runs("CH Gayle")
```
