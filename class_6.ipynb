```python
# Merging,Joining and Concatenating
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
courses = pd.read_csv("courses.csv")
matches = pd.read_csv("matches.csv")
month_1 = pd.read_csv("reg-month1.csv")
month_2 = pd.read_csv("reg-month2.csv")
deliveries = pd.read_csv("deliveries.csv")
students = pd.read_csv("students.csv")
```

```python
# pd.concat
# ignore_index
# multiindex ---> fetch using iloc
# concat dataframes horizontally
```

```python
# pd.concat()
```

```python
pd.concat([month_1,month_2])
```

```python
regs = pd.concat([month_1,month_2],ignore_index=True)
regs
```

```python
# multiIndex DataFrame
```

```python
pd.concat([month_1,month_2],keys=["month_1","month_2"])
# MultiIndex DataFrame
```

```python
# Concat two DataFrame Horizontally
```

```python
pd.concat([month_1,month_2],axis=1)
```

```python
# Merge
```

```python
# Inner-join
```

```python
students.merge(regs,how="inner",on="student_id")
```

```python
# left-join--> inner join + remaining in  left table
```

```python
temp_df = pd.DataFrame(
    {
        "student_id":[26,27,28],
        "name":["hasnain","fahad","cristiano"],
        "partner":[28,26,17]
    }
)
students = pd.concat([students,temp_df],ignore_index=True)
```

```python
students.merge(regs,how="left",on="student_id")
```

```python
# right-join--> inner join+ remaining in right table 
```

```python
students.merge(regs,how="right",on="student_id")
```

```python
# outer-join / full-join ---> inner join+ remaining in both left and right table.
```

```python
students.merge(regs,how="outer",on="student_id")
```

```python
# find total revenue generated
revenue_generated = regs.merge(courses,how="inner",on="course_id")["price"].sum()
revenue_generated
```

```python
# find month by month revenue
```

```python
month_1_revenue = month_1.merge(courses,how='inner',on='course_id')['price'].sum()
month_1_revenue
```

```python
month_2_revenue = month_2.merge(courses,how='inner',on='course_id')['price'].sum()
month_2_revenue
```

```python
# Another way(prefer)
months_df = pd.concat([month_1,month_2],keys=["month_1","month_2"]).reset_index()
res = months_df.merge(courses,how='inner',on='course_id').groupby("level_0")['price'].sum()
res
```

```python
# print registration table
# cols ---> name,course,price
```

```python
temp = regs.merge(courses,how='inner',on='course_id')
result = temp.merge(students,how='left',on="student_id")[["name","course_name","price"]]
result
```

```python
# plot bar chart for revenue/courses
result = regs.merge(courses,how="inner",on="course_id").groupby("course_name")['price'].sum().reset_index()
result.set_index("course_name",inplace=True)
result
```

```python
ax = result.plot(kind='bar', figsize=(9, 5))
ax.set_xlabel("Courses")
ax.set_ylabel("Revenue")
ax.set_title("Revenue Generated From Each Course")
```

```python
# find students who enrolled in both the month
students_enrolled_in_both_months = month_1.merge(month_2,how="inner",on="student_id").merge(students,how="inner",on="student_id").drop_duplicates(subset=["name"])["name"]
students_enrolled_in_both_months
```

```python
# print student name, partner name for all enrolled students
```

```python
enrolled_students = students[students["student_id"].isin(regs["student_id"])]
enrolled_students
```

```python
enrolled_students.reset_index(drop=True,inplace=True)
for i in range(len(enrolled_students["partner"])):
    partner_id = enrolled_students.loc[i,"partner"]
    if partner_id in enrolled_students["student_id"].values:
        partner_name = enrolled_students.loc[enrolled_students["student_id"] == partner_id,"name"].values[0]
        enrolled_students.loc[i,"partner_name"] = partner_name
    else:
        enrolled_students.loc[i,"partner_name"] = np.nan
enrolled_students
```

```python
# self join
students.merge(students,how="inner",left_on="partner",right_on="student_id")
# how left_on="partner" and right_on="student_id"
```

```python
# find top 3 stadiums with most sixes/match ratio
```

```python
# first we merge two datasets(matches,deliveries)
merged_dataset = deliveries.merge(matches,how="inner",left_on="match_id",right_on="id")
runs_only_six_df = merged_dataset[merged_dataset["batsman_runs"] == 6]
venue_groupby_object = runs_only_six_df.groupby("venue")
```

```python
def stadium_with_most_sixes(group):
    no_of_sixes_per_stadium = group["batsman_runs"].count()
    no_of_matches_per_stadium = len(group["match_id"].unique())
    result = no_of_sixes_per_stadium/no_of_matches_per_stadium
    return result
```

```python
venue_groupby_object.apply(stadium_with_most_sixes).sort_values(ascending=False).head(3)
```

```python
# pllotting(No Of Sixes per match in each stadium)
```

```python
res = venue_groupby_object.apply(stadium_with_most_sixes)
```

```python
ax = res.plot(kind='bar', figsize=(9, 5))
ax.set_xlabel("Stadiums")
ax.set_ylabel("sixes_per_match")
ax.set_title("No Of Sixes per match in each stadium")
```

```python
# find orange cap holder of all the seasons
```

```python
season_groupby_object = merged_dataset.groupby(["season","batsman"])
result = season_groupby_object["batsman_runs"].sum().reset_index().sort_values("batsman_runs",ascending=False).drop_duplicates(subset="season",keep="first").sort_values("season")
```

```python
orange_cap_holder = result[["season","batsman"]]
orange_cap_holder
```
